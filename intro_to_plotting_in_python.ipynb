```python
import numpy as np
import matplotlib
```

```python
import matplotlib.pyplot as pl
%matplotlib inline                 #We'll look at this in more detail a little later
```

- Let's make a fake timeseries and plot the cosine function using NumPy and Matplotlib

```python
times = np.linspace(0, 100, 500)
freq = 50 #Hz

#Here we use the numpy "cos" function
amplitude = np.cos(2 * np.pi * freq * times)
```

- Let's make our first call to the matplotlib pyplot subpackage and plot the above data

```python
pl.plot(times, amplitude)
pl.show()
```

```python
pl.plot(times, amplitude)

pl.xlabel("Time")
pl.ylabel("Amplitude")

pl.title("Cosine wave")

pl.show()
```

```python
pl.plot(times, amplitude)

pl.xlabel("Time", fontsize = 22)
pl.ylabel("Amplitude", fontsize = 22)

pl.title("Cosine wave", fontsize = 22)

pl.show()
```

```python
scale = -0.1

amplitude_2 = np.exp(scale * times)
```

```python
pl.plot(times, amplitude)
pl.plot(times, amplitude_2)

pl.xlabel("Time", fontsize = 22)
pl.ylabel("Amplitude", fontsize = 22)

pl.title("Two functions", fontsize = 22)

pl.show()
```

```python
#Let's add in labels to the lines:
pl.plot(times, amplitude, label = 'cosine wave')
pl.plot(times, amplitude_2, label = 'exponential function')

pl.xlabel("Time", fontsize = 22)
pl.ylabel("Amplitude", fontsize = 22)

pl.title("Two functions", fontsize = 22)

#Note this line
pl.legend(loc = 'best')

pl.show()
```
